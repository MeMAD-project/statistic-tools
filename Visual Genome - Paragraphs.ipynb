```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def combined_description_and_image(data, idx):
    # Combine descriptions:
    image_regions = [region['phrase'] for region in data[idx]['regions']]
    
    image_id = data[idx]['regions'][0]['image_id']
    
    img_path = '/scratch/cs/imagedb/picsom/databases/visualgenome/download/1.0/images/' + str(image_id) + '.jpg'
    img = mpimg.imread(img_path)
    
    plt.figure(figsize = (20, 15))
    plt.imshow(img)
    plt.show()
    
    print(f"Dense captions: \n\n{image_regions}")
    
def combined_paragraph_dense_captions_and_image(region_data, paragraph_data, paragrpah_idx):
    par = paragraph_data[paragrpah_idx]
    
    image_id = par['image_id']
    region_idx = None
    
    # find image with right id:
    
    for i, image in enumerate(region_data):
        if image['regions'][0]['image_id'] == image_id:
            print(f"Found at idx {i}")
            region_idx = i
            break
    
    print(f"Full caption: \n\n{par['paragraph']} \n")
    
    combined_description_and_image(region_data, region_idx)
    
```

```python
combined_paragraph_dense_captions_and_image(data, paragraph_data, 3)
```
